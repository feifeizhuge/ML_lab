```python
from numpy import*
from numpy.random import *
import matplotlib.pyplot as plt
%matplotlib inline
Rminus = normal(20,10,10000)        # 標準正規分布で乱数を1万個生成
plt.hist(Rminus, bins=100)   # 100本のヒストグラムを作成
plt.show()              # グラフを表示
```

```python
rminus = 20
stabw = 10
#Nehmen Sie für ein gegebenes d eine fixe Schwelle z(d) an, die genau zwischen den Mittelwerten der beiden Verteilungen liegt.
d = 0
z = zeros(10)
relkor = zeros(10)
for i in range(10):
    d = d+1
    rplus = d*stabw + rminus
    Rplus = normal(rplus,10,10000)
    z[i] = (rplus+rminus)/2
    a = len(Rminus[where( Rminus > z[i] )])
    b = len(Rplus[where( Rplus < z[i] )])
    relkor[i] = (20000-a-b)/20000
    print("Die zahl von false negaitve: ",a)
    print("Die zahl von false positve: ",b)
    print("relative Anzahl an korrekten Ergebnissen:", relkor[i])
    plt.hist(Rplus, bins=100)
    plt.hist(Rminus, bins=100)  
    plt.show() 
print(z)
```

```python
x = arange(10).reshape(10)
x = x
    
plt.plot(x, relkor)  
```

```python
p1 = np.random.randn(10000)
p2 = np.random.randn(10000) +2
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(p1, 100)
plt.title("Histgram of N(µ = 0; std = 1)")
plt.subplot(1,2,2)
plt.hist(p2, 100)
plt.title("Histgram of N(µ = 2; std = 1)")
```

```python
plt.hist(p1, 100)
plt.hist(p2, 100)
```

```python
f= np.arange(11)-4
P1 = np.append(p1.reshape(1,len(p1)),np.zeros((1,len(p1))),0)
P2 = np.append(p2.reshape(1,len(p2)),np.ones((1,len(p2))),0)
P = np.append(P1,P2,1)
P =P.T[np.argsort(P[0])]
ptrue = np.array(P[:,1])
roc = np.zeros((2,len(f)))
for i in range(len(f)):
    pred =np.zeros(len(P))
    pred[np.where(P[:,0]>f[i])]=1
    roc[1][i] = ptrue[np.where(pred==1)].sum()/10000
    roc[0][i] = pred[np.where(ptrue==0)].sum()/10000
ROC=roc[:,np.argsort(roc[0])]
```

```python
plt.plot(ROC[0],ROC[1], label='Analtical')
plt.legend()
plt.xlabel("FP rate")
plt.ylabel("TP rate")
plt.title("ROC curve")
```
